```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
string_columns = list(df.dtypes[df.dtypes == 'object'].index)
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
#Missing value

missing = df.isna().sum().to_frame("Missing Values")

missing["Percentage"] = (missing["Missing Values"] / len(df) * 100).round(2)

missing.sort_values("Missing Values", ascending=False)
```

```python
df["issue_reported_at"] = (
    df["issue_reported_at"]
      .str.replace("_", " ", regex=False)
)

df["issue_responded"] = (
    df["issue_responded"]
      .str.replace("_", " ", regex=False)
)
```

```python
df[['issue_reported_at','issue_responded']].head(10)
```

```python
df["issue_reported_at"] = pd.to_datetime(
    df["issue_reported_at"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)

df["issue_responded"] = pd.to_datetime(
    df["issue_responded"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)
```

```python
df["issue_reported_at"].isna().sum()
```

```python
df["issue_responded"].isna().sum()
```

```python
df["response_time_minutes"] = (
    df["issue_responded"] -
    df["issue_reported_at"]
).dt.total_seconds() / 60
```

```python
df["response_time_minutes"].describe()
```

```python
# Number of negative response times
negative_count = (df["response_time_minutes"] < 0).sum()

# Percentage of dataset
negative_percentage = (
    negative_count / len(df)
) * 100

print(negative_count)
print(round(negative_percentage, 2))
```

```python
negative_rows = df.loc[
    df["response_time_minutes"] < 0,
    [
        "issue_reported_at",
        "issue_responded",
        "response_time_minutes"
    ]
]

negative_rows.head(20)
```

```python
df["customer_city"] = df["customer_city"].fillna("Unknown")

df["product_category"] = df["product_category"].fillna("Unknown")
```
